1. Scrape the details of most viewed videos on YouTube from Wikipedia:

```python
import pandas as pd
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
from bs4 import BeautifulSoup
from selenium.common.exceptions import StaleElementReferenceException,NoSuchElementException,ElementNotInteractableException
import re
import requests
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
# opening the chrome
driver=webdriver.Chrome()
```

```python
Rank=[]
Name=[]
Artist=[]
Views=[]
Date=[]
```

```python
#Scrapping the Name
try:
    name=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"][1]/tbody/tr/td[1]')
    for i in name:
        Name.append(i.text)

except NoSuchElementException:
    Name.append("Not Present")

Name
```

```python
# Scrapping the artist
try:
    artist=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"][1]/tbody/tr/td[2]')
    for i in artist:
        Artist.append(i.text)

except NoSuchElementException:
    Artist.append("Not Present")

Artist
```

```python
# Scrapping the views
try:
    views=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"][1]/tbody/tr/td[3]')
    for i in views:
        Views.append(i.text)

except NoSuchElementException:
    Views.append("Not Present")

Views
```

```python
# Scrapping the date of video upload
try:
    date=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"][1]/tbody/tr/td[4]')
    for i in date:
        Date.append(i.text)

except NoSuchElementException:
    Date.append("Not Present")

Date
```

```python
Yt=pd.DataFrame({"Video_Name":Name,"Uploader":Artist,"Views_In_Billions":Views,"Video_Upload_Date":Date})
Yt
```

8. Details of Datasets from UCI machine learning repositories.
Url = https://archive.ics.uci.edu/ You
have to find the following details:
A) Dataset name
B) Data type
C) Task
D) Attribute type
E) No of instances
F) No of attribute G) Year
 Note: - from the home page you have to go to the Show All Dataset page through code

```python
driver=webdriver.Chrome()
driver.get(" https://archive.ics.uci.edu/")
```

```python
# Clicking the dataset
click=driver.find_element(By.XPATH,'//a[@class="btn btn-primary"]')
click.click()
```

```python
# scrapping url
url=[]
dataset=driver.find_elements(By.XPATH,'//a[@class="link-hover link text-xl font-semibold"]')
for i in dataset:
    url.append(i.get_attribute('href'))

next_btn=driver.find_element(By.XPATH,'//div[@class="btn-group"]/button[2]')
```

```python
len(url)
```

```python
url
```

```python
# creating empty variables
Dataset_Name=[]
Data_Type=[]
Task=[]
Attribute_Type=[]
No_Of_Instances=[]
No_Of_Attribute=[]
Year=[]
```

```python
# scraping dataset name
for i in url:
    driver.get(i)
    time.sleep(2)
    try:
        dataset=driver.find_element(By.XPATH,'//div[@class="flex w-full items-center justify-between gap-4 bg-primary p-2"]/div/div/div')
        Dataset_Name.append(dataset.text)
    except NoSuchElementException:
        Dataset_Name.append("NA")

    try:
        type=driver.find_element(By.XPATH,'//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[1]/p')
        Data_Type.append(type.text)

    except NoSuchElementException:
        Data_Type.append("NA")

    try:
        task=driver.find_element(By.XPATH,'//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[3]/p')
        Task.append(task.text)

    except NoSuchElementException:
        Task.append("NA")

    try:
        attribute=driver.find_element(By.XPATH,'//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[4]/p')
        Attribute_Type.append(attribute.text)

    except NoSuchElementException:
        Attribute_Type.append("NA")

    try:
        noi=driver.find_element(By.XPATH,'//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[5]/p')
        No_Of_Instances.append(noi.text)

    except NoSuchElementException:
        No_Of_Instances.append("NA")

    try:
        noa=driver.find_element(By.XPATH,'//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[6]/p')
        No_Of_Attribute.append(noa.text)

    except NoSuchElementException:
        No_Of_Attribute.append("NA")

    try:
        year=driver.find_element(By.XPATH,'//h2[@class="text-sm text-primary-content"]')
        Year.append(year.text)

    except NoSuchElementException:
        Year.append("NA")

       
```

```python
print(len(Dataset_Name),len(Data_Type),len(Task),len(Attribute_Type),len(No_Of_Instances),len(No_Of_Attribute),len(Year))
```

```python
data={"Dataset_Name":Dataset_Name,"Data_Type":Data_Type,"Task":Task,"Attribute_Type":Attribute_Type,"No_Of_Instances":No_Of_Instances,
      "No_Of_Attribute":No_Of_Attribute,"Year":Year}
df=pd.DataFrame(data)
df
```
